```python
import trectools
```

```python
#defining paths for Qrels and results file
qrels_path = "./cranqrel.trec.txt"
results_path = "./output_vsls.txt"

#making qrels and run objects
qrels = trectools.TrecQrel(qrels_path)
results = trectools.TrecRun(results_path)

#making trec eval object and getting scores
te = trectools.TrecEval(results, qrels)
map_score = te.get_map()
precision = te.get_precision(depth=5)
ndcg_score = te.get_ndcg()
```

```python
print("MAP:", round(map_score,5))
print("P@5:", round(precision,5))
print("NDCG:", round(ndcg_score,5))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Models
models = ['VSM', 'BM25', 'UNIGRAM']

# Precision scores for the first set of scores
precision_scores_set1 = [0.0071, 0.0053, 0.0062]

# Precision scores for the second set of scores
precision_scores_set2 = [0.0088, 0.008, 0.008]

# Set the width of the bars
bar_width = 0.35

# Set the position of the bars on the x-axis
r1 = np.arange(len(models))
r2 = [x + bar_width for x in r1]

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(r1, precision_scores_set1, color='skyblue', width=bar_width, edgecolor='grey', label='Models with Stemming')
plt.bar(r2, precision_scores_set2, color='salmon', width=bar_width, edgecolor='grey', label='Models with Lemmatization')

# Add xticks on the middle of the group bars
plt.xlabel('Information Retrieval Models', fontweight='bold')
plt.xticks([r + bar_width/2 for r in range(len(models))], models)

# Add y label
plt.ylabel('P@5 Score', fontweight='bold')

# Add title
plt.title('Comparison of Precision Scores for Different Models', fontweight='bold')

# Add legend
plt.legend()



# Show plot
plt.show()
```
